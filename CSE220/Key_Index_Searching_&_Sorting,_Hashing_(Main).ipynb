```python
# Task 1 on Key index Searching & Sorting

class KeyIndex:
  def __init__(self,l):
    self.lst = l

  def maxi(self):
    mx_n = self.lst[0]
    for i in range(len(self.lst)):
      if mx_n < self.lst[i]:
        mx_n = self.lst[i]
    return mx_n

  def mini(self):
    mn_n = self.lst[0]
    for i in range(len(self.lst)):
      if mn_n > self.lst[i]:
        mn_n = self.lst[i]
    return mn_n

  def aux(self):

    arr = [0]*(self.maxi()+abs(self.mini())+1)
    for i in range(len(self.lst)):
      arr[self.lst[i]+abs(self.mini())] += 1
    return arr
  
  def search(self, val):

    # print("Maximum: ", self.maxi())
    # print("Minimum: ", self.mini())

    if self.mini() > 0:
      for i in range(len(self.aux())):
        if self.aux()[i] > 0:
          if i - self.mini() == val:
            return True
      return False
    else:
      for i in range(len(self.aux())):
        if self.aux()[i] > 0:
          if i + self.mini() == val:
            return True
      return False

  def sort(self):

    if self.mini() > 0:
      k = 0
      sorted_arr = self.aux() 
      for i in range(len(sorted_arr)):
        if sorted_arr[i] > 0:
          for j in range(sorted_arr[i]):
            self.lst[k] = i-self.mini()
            k += 1
      return self.lst
    else:
      k = 0
      sorted_arr = self.aux() 
      for i in range(len(sorted_arr)):
        if sorted_arr[i] > 0:
          for j in range(sorted_arr[i]):
            self.lst[k] = i+self.mini()
            k += 1
      return self.lst


l = [2,-4,3,5,8,1,3,4]
h = KeyIndex(l)
print(h.search(-3))
print(h.sort())
```
