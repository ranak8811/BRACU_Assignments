```python
# task-22

class Hotel:
  def __init__(self,hn):
    self.hn = hn
    self.sl = []
    self.sid = 0
    self.gl = []
    self.gid = 0
    self.sd = {}
    self.gd = {}

  def addStuff(self,*sinfo):
    self.sl = list(sinfo)
    self.sid += 1
    print(f'Staff With ID {self.sid} is added')

  def getStuffById(self,id):
    self.sd[id] = self.sl
    self.sphn = '000'
    return (f'Staff ID: {id}\nName: {self.sl[0]}\nAge: {self.sl[1]}\nPhone no.: {self.sphn}')

  def addGuest(self,*ginfo):
    self.gl = list(ginfo)
    self.gid += 1
    print(f'Guest With ID {self.gid} is created')

  def getGuestById(self,id):
    self.gd[id] = self.gl
    return (f'Guest ID: {id}\nName: {self.gl[0]}\nAge: {self.gl[1]}\nPhone no.: {self.gl[2]}')

  def allStaffs(self):
    print('All Staffs:')
    print('Number of Staff: ',self.sid)
    for k,v in self.sd.items():
      print(f'Staff ID: {k} Name: {v[0]} Age: {v[1]} Phone no: {self.sphn}')

  def allGuest(self):
    print('All Guest:')
    print('Number of Guest: ',self.gid)
    for k,v in self.gd.items():
      print(f'Staff ID: {k} Name: {v[0]} Age: {v[1]} Phone no: {v[2]}')




# Write your code here
# Do not change the following lines of code.
h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest('Carol',35,'123')
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, '431')
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()


print(h.sd)
print(h.gd)

print(h.sl)
print(h.gl)
```
